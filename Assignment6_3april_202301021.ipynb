```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('test.csv')
print(df)
```

```python
x=df["x"]
print(x)
y=df["y"]
print(y)
```

```python
data=list(zip(x,y))
data
```

```python
kmeans=KMeans(n_clusters=4)
kmeans.fit(data)
plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
